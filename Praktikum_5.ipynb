```python

# Materi
my_list = [1,2,3,4,5]

# append untuk menambahkkan index
my_list.append(6)
print(f"nambah",my_list)

# insert untuk menyisipkan index
my_list.insert(2,9)
print(f"selip",my_list)

# remove untuk menghapus index
my_list.remove(3)
print("hapus",my_list)

# popped_value untuk menghapus elemen pada index dan menyimpannya dalam variable
angka = my_list.pop(4)
print(my_list)
print(angka)

# Untuk print huruf tetrentu pada array
ar_ket = ['first', 'world', 'walk']
print("ambil angka :",ar_ket[2][2])


#========================================================
buah1 = ['anggur', 'melon', 'apel']
buah2 = ['semangka', 'jeruk', 'pisang']

# menggabungkan var buah1 dan var buah2
buahbuahan = buah1 + buah2
print("nambah buah",buahbuahan)

# melakukan perulangan pada var buah1 sebanyak 3x
ngulangbuah = buah1 * 3
print(f"ngulang buah",ngulangbuah)

# =======================================================
# tipe datanya tentuin dulu
value = int(input('Masukkan nilai atau string: '))

match value:
    case 1 | 2 | 3:
      print('ini angka 1,2 atau 3')
    case 'apple' | 'banana':
        print("Mencocokkan 'apple' atau 'banana'")
    case _:
        print("Default case")

# ======================================================
pemrograman = input('bahasa pemrograman yang inginn kamu pelajari ')

match pemrograman:
  case'javascript':
    print('kamu bisa menjadi web developer')
  case'python':
    print('kamu bisa menjadi data analis')
  case'php':
    print('kamu bisa menjadi backend developer')
  case'java':
    print('kamu bisa menjadi mobile app developer')



# ==========================================================================================================================================
# Soal
# 1. Buat variabel list dengan value : [namaKendaraan, JenisKendaraan, ccKendaraan, warna kendaraan, roda kendaraan] tambahkan dari list tersebut di belakang dengan value : [harga kendaraan, tipe kendaraan] tambahkan setelah jenis kendaraan dengan value [merk kendaraan]
# 2. Buat program python dengan match case untuk menghitung luas bangun datar : jika pilih 1, maka menghitung luas persegi jika pilih 2, maka menghitung luas lingkaran jika pilih 3, maka menghitung luas segitiga selain pilihan di atas, maka keterangan : salah pilih angka

# Soal No 1
ar_kendaraan = ['Honda Civic', 'Mobil', '125cc', 'Biru', '4 roda']
print(ar_kendaraan)

# Menambahkan harga kendaraan dan tipe kendaraan
# Menambahkan harga kendaraan
ar_kendaraan.append('586jt')
print(ar_kendaraan)

# Menambahkan tipe kendaraan
ar_kendaraan.append('Civic')
print(ar_kendaraan)

# Untuk menyisipkan merk motor ke dalam data list di index 2
ar_kendaraan.insert(2,'Honda')
print(ar_kendaraan)

# Soal 2
ket = '''selamat datang di aplikasi menghitung luas bangun datar
masukan pilihan:
1. untuk menghitung luas persegi
2. untuk menghitung luas lingkaran
3. untuk menghitung luas segitiga
'''

pilihan = input (ket)

match pilihan:
    case "1":
        print ("kamu memilih 1 untuk menghitung luas persegi")
        sisi = int(input ("masukan sisi ?"))
        luasP = sisi * sisi
        print ("luas persegi yang sisinya ", sisi, " adalah ",luasP)
    case "2":
        print ("kamu memilih 2 untuk menghitung luas lingkaran")
        jari2 = float(input ("masukan jari2 ?"))
        luasL = 3.14 * jari2 * jari2
        print ("luas lingkaran yang jari2nya ", jari2, " adalah ",int(luasL))
    case "3":
        print ("kamu memilih 3 untuk menghitung luas segitiga")
        alas = int(input ("masukan alas ?"))
        tinggi = int(input ("masukan tinggi ?"))
        luasS = 0.5 * alas * tinggi
        print ("luas segitiga yang alasnya ", alas, " dan tingginya",tinggi , " adalah ",int(luasS))
    case _:
        print("salah memilih pilihan")
```
